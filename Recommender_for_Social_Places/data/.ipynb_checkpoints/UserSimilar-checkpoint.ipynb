```python
import pandas as pd  
from sqlalchemy import create_engine  
import pymysql
import csv
import os
from surprise import Dataset
from surprise import accuracy
from surprise.model_selection import train_test_split
from surprise import KNNBasic
from surprise import BaselineOnly
from surprise import Reader
#connect to mysql

path_upper = os.path.abspath(os.path.dirname(os.getcwd()))
#load mysql information
f = open(path_upper+"/config.txt")
for line in f:
    info = line.split("=")
    if info[0] == "MYSQL_USER":
        mysql_user = info[1]
    elif info[0] == "MYSQL_PASSWORD":
        mysql_password = info[1]
    elif info[0] == "MYSQL_DB_NAME":
        mysql_db_name = info[1]
config = {
          'host':'127.0.0.1',
          'port':3306,
          'user':mysql_user,
          'password':mysql_password,
          'db':mysql_db_name,
          'charset':'utf8mb4',
          'cursorclass':pymysql.cursors.DictCursor,
          }

db = pymysql.connect(**config)

cursor = db.cursor()

cursor.execute('select * from preference')
result = cursor.fetchall()
cursor.close() 
df=pd.DataFrame(result)

df_length = df.shape[0]
user =[]
item = []
rating =[]

item_name = ['American', 'Asian', 'Halal', 'Indian', 'Mexican', 'bars', 'coffee', 'museum', 'park', 'shopping']
item_list =[]
for i in item_name:
    item_list.append(df[i].values.tolist())
for ratings in item_list:
    for rate in ratings:
        rating.append(int(rate))


user_list =df['username'].values.tolist()

for i in item_name:
    user.extend(user_list)
    for j in user_list:
        item.append(i)
        




#load into mid_ouput.csv
dataframe = pd.DataFrame({'user':user,'item':item,'rating':rating})
dataframe.to_csv("mid_output.csv",index=False,sep=',')


ratings_dict = {'user': user,
                'item': item,
                'rating': rating}
df = pd.DataFrame(ratings_dict)

# A reader is still needed but only the rating_scale param is requiered.
reader = Reader(rating_scale=(0, 1))

# The columns must correspond to user id, item id and ratings (in that order).
data = Dataset.load_from_df(df[['user', 'item', 'rating']], reader)



all_trainset = data.build_full_trainset()
algo = KNNBasic(k=40,min_k=1,sim_options={'user_based': True}) # sim_options={'name': 'cosine','user_based': True} cosine/msd/pearson/pearson_baseline
algo.fit(all_trainset)
def getSimilarUsers(top_k,u_id):
    user_inner_id = algo.trainset.to_inner_uid(u_id)
    user_neighbors = algo.get_neighbors(user_inner_id, k=top_k)
    user_neighbors = [algo.trainset.to_raw_uid(inner_id) for inner_id in user_neighbors]
    return user_neighbors


sim_user_dic = {}
for u in user_list:
    sim_list = getSimilarUsers(5,u)
    sim_user_dic[u]=sim_list

with open('user_output.csv', 'w',newline='') as csv_file:
    writer = csv.writer(csv_file)
    for key, value in sim_user_dic.items():
        writer.writerow([key, value])
```
